**Automatização de entrada de dados no sistema.**

Não pude trazer o código rodando no PC da empresa para não comprometer os dados da empresa.
Compreendo também que a entrada destes dados poderiam ser feitos de uma maneira mais eficaz
como por exemplo: Consumir a API do GoogleSheets e inputar as informações diretamente no banco de dados
da empresa. Mas para isso me faltou ferramentas a primeira é que para consumir a API do GoogleSheets teria
que pagar, segundo que não me foi disponibilizado acesso ao banco de dados da empresa.

*Portanto o código abaixo é uma ilustração do código real
*Para ilustrar iremos usar o GoogleSheets

```python
import pandas as pd
import pyautogui
```

```python
arquivo = pd.read_csv('C:/Users/User/Desktop/Inspecao - Check-list.csv')
arquivo.head()
```

```python
#Acessando o Local
pyautogui.alert("O código vai começar. Não mexa no PC")
pyautogui.hotkey('win','r')
pyautogui.write('chrome')
pyautogui.press('enter')
pyautogui.sleep(0.5)
pyautogui.write('https://docs.google.com/spreadsheets/d/1vmzLUawZgrTXDSLhyZ1iPXxZWcW8q4TwafujWTB_esw/edit#gid=0')
pyautogui.press('enter')
pyautogui.sleep(2)
pyautogui.press('down')
#Loop
for i in range(6):
    pyautogui.typewrite(arquivo["ID"][i])
    pyautogui.press('right')
    pyautogui.typewrite(arquivo["Motorista"][i])
    pyautogui.press('right')
    pyautogui.typewrite(arquivo["Placa"][i])
    pyautogui.press('right')
    pyautogui.typewrite(arquivo["Observações"][i])
    pyautogui.press('right')
    pyautogui.press('down')
    pyautogui.press('home')
```
